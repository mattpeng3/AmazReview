```python
import pandas as pd
import operator
```

```python
#pandas to read csv file
mobile = pd.read_csv("phones.csv")
```

```python
del mobile["Product Name"]
del mobile["Brand Name"]
del mobile["Price"]
del mobile["Review Votes"]


mobile_five = mobile.loc[mobile['Rating'] == 5]
mobile_four = mobile.loc[mobile['Rating'] == 4]
mobile_three = mobile.loc[mobile['Rating'] == 3]
mobile_two = mobile.loc[mobile['Rating'] == 2]
mobile_one = mobile.loc[mobile['Rating'] == 1]
```

```python
commonWords = ['the', 'be', 'to', 'of', 'and', 'a', 'in', 'that', 'have',
              'i', 'it', 'for', 'not', 'on', 'with', 'he', 'as', 'you',
              'do', 'at', 'this', 'but', 'his', 'by', 'from', 'they', 'we',
              'say', 'her', 'she', 'or', 'will', 'an', 'my', 'one', 'all',
              'would', 'there', 'their', 'what', 'so', 'up', 'out', 'if',
              'about', 'who', 'get', 'which', 'go', 'when', 'me', 'make',
              'can', 'like', 'time', 'no', 'just', 'him', 'know', 'take',
              'person', 'into', 'year', 'your', 'good', 'some', 'could',
              'them', 'see', 'other', 'than', 'then', 'now', 'look', 'only',
              'come', 'its', "it's", 'over', 'think', 'also', 'back', 'after',
              'use', 'two', 'how', 'our', 'work', 'first', 'well', 'way', 'even',
              'new', 'want', 'because', 'any', 'these', 'give', 'day', 'most', 'us',
               'who'
    ]
```

```python
#### deletes the symbols out of the words for instance such as . and '
def del_sym(txt):
    del_symbols = ['.', "'", "!", "@", "#", "$", "%", "^", "&", "*", "(", ")", "{", "}",
                       "1", "2", "3", "4", "5", "6", "7", "8", "9", ":", ";", "<", ">", "?",
                       ",", "[", "]", "-", "_", "+", "="
                  ]
     
    i = 0
    while i < len(txt):
        if txt[i] in del_symbols:
            txt = txt[0:i] + txt[i+1:]
            i-=1
        i+=1
    return txt
    

#uncomment below to test function
#word = "hi''.a"
#words = "he344o&*o"
#del_sym(word)  
#del_sym(words)                
    
```

```python
def addText(dataTable):
    sentence = ""
    for i in range(dataTable.shape[0]):
        word = str(dataTable.iloc[i]["Reviews"])
        sentence += " " + word
    return sentence

def deleteCommon(str):
    sentence = ""
    conden = str.split(" ")
    for i in range(len(conden)):
        if conden[i].lower() not in commonWords:
            sentence += " " + conden[i]
    return sentence

def rewriteText(dataTable):
    allData = deleteCommon(addText(dataTable))
    allDataSplit = allData.split(" ")
    d = {}
    for i in range(len(allDataSplit)):
        if allDataSplit[i] not in d:
            d[allDataSplit[i]] = 1
        else:
            d[allDataSplit[i]] += 1
            
    sorted_d = sorted(d.items(), key=operator.itemgetter(1), reverse=True)
    return sorted_d


one_text = rewriteText(mobile_one)
two_text = rewriteText(mobile_two)
three_text = rewriteText(mobile_one)
four_text = rewriteText(mobile_four)
five_text = rewriteText(mobile_five)
```

```python
##Cleans the words using the delete symbol function above
def clean(arr):
    array = []
    for word in arr:
        array.append(del_sym(str(word))) 
    return array
```

```python
##word_ratings creates array of words associated with each rating
##num paramteter is the top whatsoever words
def word_ratings(txt, num):
    words = []
    i=0
    while i < num:     
        words.append(txt[i][0])
        i+=1
    return words
```

```python
one_star = word_ratings(one_text, 500)
two_star = word_ratings(two_text, 500)
three_star = word_ratings(three_text, 500)
four_star = word_ratings(four_text, 500)
five_star = word_ratings(five_text, 500)
```

```python
one = clean(one_star)
two = clean(two_star)
three = clean(three_star)
four = clean(four_star)
five = clean(five_star)
```

```python
#another solution to solving getting rid of common/nonsignificant 
#words that do not have a positive or negative connotation
import matplotlib.pyplot as plt

def full_words(txt):
    array = []
    for i in txt:
        if len(i) == 2:
            first = del_sym(str(i[0]))
            second = i[1]
            new_item = (first, second)
            array.append(new_item)
    return array

test_one = full_words(one_text)
test_two = full_words(two_text)
test_three = full_words(three_text)
test_four = full_words(four_text)
test_five = full_words(five_text)
```

```python
########The purpose of this function and the above function is basically to create a array of tuples from the original
########array of tuples but this new array has now cleaned up all the symbols for each word
########The below function creates a dictionary of the form {word length: # words with that word length}
########Example: {5: 100} means 100 words have 5 letters

def create_count(array):
    d = {}
    for i in array:
        if len(i[0]) in d:
            d[len(i[0])] += i[1]
        else:
            d[len(i[0])] = i[1]
    return d


##Below are dictionaries where the value is length of word and key is the number of words in the review text with that
##letter count
t_one = create_count(test_one)
t_two = create_count(test_two)
t_three = create_count(test_three)
t_four = create_count(test_four)
t_five = create_count(test_five)
```

```python
##this function, from the t_one... etc variables, grabs only the words of length x(any length) 
##that occur more than y times 
def filt(dic, y):
    d = {}
    for x in dic:
         if dic[x] > y:
                d[x] = dic[x]
    return d
```

```python
def point_value(review, one, two, three, four, five):
    value = 0
    words = review.split()
    length = len(words)
    best_words = []
    worst_words = []
    best_value = 0
    worst_value = 10
    indicator = False
    
    def best_worst(val):
        nonlocal best_value
        nonlocal worst_value
        nonlocal indicator
        
        if val >= best_value:
            best_value = val
            indicator = 1
            return True
        elif val <= worst_value:
            worst_value = val
            indicator = 0
            return True 
        return False
    
    def common_counter(word):
        common_words = 0
        if word in one:
            common_words+=1
        if word in two:
            common_words+=1
        if word in three:
            common_words+=1
        if word in four:
            common_words+=1
        if word in five:
            common_words+=1
        return common_words
    
    for i in range(length):
        common_count = common_counter(words[i])
        if words[i] in one and common_count < 3:
            value += 1
            if best_worst(1):
                if indicator:
                    best_words.append(words[i])
                else:
                    worst_words.append(words[i])
                    
        elif words[i] in two and common_count < 3:
            value += 2
            if best_worst(2):
                if indicator:
                    best_words.append(words[i])
                else:
                    worst_words.append(words[i])
                    
        elif words[i] in three and common_count < 3:
            value += 3
            if best_worst(3):
                if indicator:
                    best_words.append(words[i])
                else:
                    worst_words.append(words[i])
                    
        elif words[i] in four and common_count < 3:
            value += 4
            if best_worst(4):
                if indicator:
                    best_words.append(words[i])
                else:
                    worst_words.append(words[i])
                    
        elif words[i] in five and common_count < 3:
            value += 5
            if best_worst(5):
                if indicator:
                    best_words.append(words[i])
                else:
                    worst_words.append(words[i])
                    
        else:
            length -= 1
        common_words = 0
    
    if length == 0:
        return 'review text does not have significant words'
    score = (value) / length
    
    print('score: ', score)
    print('best used words: ', best_words)
    print('worst used words: ', worst_words)
    
```
