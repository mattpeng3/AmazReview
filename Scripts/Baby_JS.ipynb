```python
#gets a dictionary of words and counts
def count_uniques(l):
    uniques= {}
    for i in l:
        if i not in uniques:
            uniques[i] = 1
        else:
            uniques[i] += 1
    return uniques
        
```

```python
def list_and_sort(dic):
    l = []
    for i in dic:
        l+= [[i,dic[i]]]
    return sorted(l,key = lambda x: x[1], reverse = True)
```

```python
def most_common(ser,n):
    #returns the n most common words
    common = []
    for i in range(len(ser)):
        common += ser[i]
    #common = 
    common = sorted(common,key = lambda x: x[1], reverse = True)
    return common[:n]
```
