```python
di = {}
```

```python
def adder(lst):
    counter = 0

    for word in lst:
        if word not in di:
            for i in range(len(lst)):
                if word == lst[i]:
                    counter += 1
            di[word], counter = counter, 0
        
       
```

```python
for lis in a:
    adder(lis.split())
```

```python
one = ['bad', 'awful', 'hate', 'gross']
two = ['not', 'horrible', 'worst']
three = ['ok', 'average', 'somewhat']
four = ['nice', 'better', 'good']
five = ['best, great', 'amazing']
```

```python
def point_value(review):
    one = ['bad', 'awful', 'hate', 'gross']
    two = ['not', 'horrible', 'worst']
    three = ['ok', 'average', 'somewhat']
    four = ['nice', 'better', 'good']
    five = ['best, great', 'amazing']
    value = 0
    words = review.split()
    length = len(words)
    best_words = []
    worst_words = []
    best_value = 0
    worst_value = 10
    indicator = False
    
    def best_worst(val):
        nonlocal best_value
        nonlocal worst_value
        nonlocal indicator
        
        if val >= best_value:
            best_value = val
            indicator = 1
            return True
        elif val <= worst_value:
            worst_value = val
            indicator = 0
            return True 
        return False
            
    
    for i in range(length):
        if words[i] in one:
            value += 1
            if best_worst(1):
                if indicator:
                    best_words.append(words[i])
                else:
                    worst_words.append(words[i])
                    
        elif words[i] in two:
            value += 2
            if best_worst(2):
                if indicator:
                    best_words.append(words[i])
                else:
                    worst_words.append(words[i])
                    
        elif words[i] in three:
            value += 3
            if best_worst(3):
                if indicator:
                    best_words.append(words[i])
                else:
                    worst_words.append(words[i])
                    
        elif words[i] in four:
            value += 4
            if best_worst(4):
                if indicator:
                    best_words.append(words[i])
                else:
                    worst_words.append(words[i])
                    
        elif words[i] in five:
            value += 5
            if best_worst(5):
                if indicator:
                    best_words.append(words[i])
                else:
                    worst_words.append(words[i])
                    
        else:
            length -= 1
        
    score = (value) / length
    
    print('score: ', score)
    print('best used words: ', best_words)
    print('worst used words: ', worst_words)
    
            
```

```python
hi = 'this product is good and amazing but somewhat and nice ok'
a = hi.split()
point_value(hi)
```
